```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
dff= pd.read_sql(QUERY, conn)
count , bins = np.histogram(dff['ConvertedComp'].dropna())
dff['ConvertedComp'].plot(kind = 'hist', figsize=(8,6))
plt.show()
```

Plot a box plot of `Age.`

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs , CodeRevHrs
FROM master
"""
df_bubb=pd.read_sql_query(QUERY, conn)
norm_age= (df_bubb['Age'] - df_bubb['Age'].min() / df_bubb['Age'].max() - df_bubb['Age'].min())
df_bubb.plot(kind = 'scatter' , x = 'WorkWeekHrs' , y ='CodeRevHrs' , alpha= 0.5 ,figsize = (7,6) ,  s = norm_age * 100 + 10 )
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY ="""
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stack= pd.read_sql_query(QUERY, conn)

df_stack.info()
```

```python
df_stack= df_stack.loc[( df_stack['Age'] >= 30) & ( df_stack['Age'] <= 35)]
df_stack
```

```python
# your code goes here
import matplotlib.pyplot as plt
df_stack.groupby(['CodeRevHrs' , 'Age'])['WorkWeekHrs'].median().sort_values(ascending = False).unstack().plot( kind = 'bar' , stacked = True , figsize =(7,6))

plt.xlabel('CodeRevHrs')
plt.ylabel('WorkWeekHrs')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT  ConvertedComp , Age
FROM master 
"""
df_line= pd.read_sql_query(QUERY, conn)
df_line.info()
```

```python
df_line= df_line.loc[( df_line['Age'] >= 45) & ( df_line['Age'] <= 60)]
df_line.dropna(inplace = True)
df_line
```

```python

df_line=df_line.groupby('Age' , axis= 0).median()


df_line
```

```python
df_line.plot(kind = 'line')
plt.title('Average Compensation Based on Age Groups')
plt.xlabel('age')
plt.show()
```
